```python
import bs4
import pandas as pd
import urllib
import re

from urllib import request
```

```python
with open("brands.txt", "r") as f:
    text = f.read()
list_brands = eval(text)
```

```python
true = 'true'
null = 'null'
false = 'false'
```

```python
all_brands = {}
i = 0
for brand in list_brands:
    i += 1
    nb = 0
    new_url = "https://www.tag-walk.com/en/look/search?designer=" + brand[1] + "&page=1"
    url = ""
    list_models = []


    while new_url != url:

        url = new_url
        request_text = request.urlopen(url).read()
        page = bs4.BeautifulSoup(request_text, "lxml")


        rows = page.find_all("div", {"class" : "photo-list-item photo-list-item--x6"})
        for row in rows:
            if row != '':
                try: 
                    name = eval(row.get("data-individuals"))[0]['name']
                    list_models.append(name)
                except:
                    nb += 1   
        
        lien = page.find('a', {"title":"Next page"})
        if lien :
            new_url = "https://www.tag-walk.com" + lien.get("href")
    
    list_no_duplicate = [[x,list_models.count(x)] for x in set(list_models)]
    list_no_duplicate.sort(key = lambda x : x[1], reverse=True)

    all_brands[brand[0]] = [list_no_duplicate, nb]
    if i % 10 == 0:
        print(i , "brands done")
    
    if i == 2:
        break
```
