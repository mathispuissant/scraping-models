```python
nb = 0
new_url = "https://www.tag-walk.com/en/look/search?designer=chanel&page=1"
url = ""
list_models = []

# extract all brands
 request_text = request.urlopen(url).read()
 page = bs4.BeautifulSoup(request_text, "lxml")
 rows = page.find_all("li", {"class" : "select2-results__options"})


while new_url != url:

    url = new_url
    request_text = request.urlopen(url).read()
    page = bs4.BeautifulSoup(request_text, "lxml")


    rows = page.find_all("div", {"class" : "photo-list-item photo-list-item--x6"})
    for row in rows:
        if row != '':
            try: 
                name = eval(row.get("data-individuals"))[0]['name']
                list_models.append(name)
            except:
                nb += 1   
    
    lien = page.find('a', {"title":"Next page"})
    if lien :
         new_url = "https://www.tag-walk.com" + lien.get("href")
```

```python
print(nb)
```

```python
list_no_duplicate = [[x,list_models.count(x)] for x in set(list_models)]
list_no_duplicate.sort(key = lambda x : x[1], reverse=True)
```

```python
with open("celebrities.txt", "r") as f:
    text = f.read()
list_celebrities = text.split("0")
```

```python
muses = []
for model in list_no_duplicate:
 #   if model[1] < 3:
#        break
    if model[0] in list_celebrities:
        muses.append(model[0])
    
```
